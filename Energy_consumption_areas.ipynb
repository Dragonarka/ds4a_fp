```python
# import libraries
import xlrd
import pandas              as pd # The gold standard of Python data analysis, to create and manipulate tables of data
import numpy               as np # The Python module for processing arrays which/Pandas is based on
import matplotlib.pyplot   as plt # The gold standard of Python data visualization, but can be complex to use
import seaborn             as sns; sns.set() # A package to make Matplotlib visualizations more aesthetic
```

```python
#import datasets
#dataset demand by operators 2008 - 2018 hourly
demand_oper = pd.DataFrame()

for y in range(2008,2019):
    data = pd.read_excel('./Demanda_por_OR/Demanda_por_OR_'+ str(y) +'.xlsx')
    data.columns = data.iloc[1]
    data = data[2:data.size]
    data['Fecha'] = pd.to_datetime(data ['Fecha'])
    #data.drop(data.index[0:3])
    demand_oper = demand_oper.append (data)
```

```python
demand_oper.head()
```

```python
# Dataset operators and their areas
oper_area_df = pd.read_excel('Operator_Area.xlsx')
```

```python
oper_area_df.head()
```

```python
demand_oper_area_df = pd.DataFrame ()
demand_oper_area_df = pd.merge(demand_oper, oper_area_df, how="left", left_on="Código Distribuidor", right_on="Operator").drop(columns=['Operator','Código Distribuidor','Version'])
demand_oper_area_df['Demand'] = demand_oper_area_df.iloc[:,1:24].sum(axis=1)
#demand_oper_area_df.sort_values(by=['Fecha'])
demand_oper_area_df.head()
```

```python
demand_sum_df = demand_oper_area_df[['Fecha','Region','Demand']]
demand_sum_df.head()
```

```python
demand_df = demand_sum_df.groupby(['Fecha','Region']).sum()
demand_df.head(50)
demand_df = demand_df.reset_index()
demand_df.set_index(['Fecha'], inplace=True)
```

```python
# Dataset maximun power daily
mpower_df = pd.DataFrame()
for y in range(2008,2019):
    data = pd.read_excel('./Demanda Potencia/Demanda_Maxima_De_Potencia_'+ str(y) +'.xlsx')
    data.columns = data.iloc[2]
    data = data[3:data.size]
    data['Fecha'] = pd.to_datetime(data ['Fecha'])
    data= data.reset_index()
    data= data[['Fecha','Demanda Máxima de Potenca kW']]
    mpower_df = mpower_df.append (data)
mpower_df.head()
```

```python
mpower_df.shape
```

```python
demand_power_df = pd.DataFrame ()
demand_power_df = pd.merge(demand_df, mpower_df, how="left",left_index = True, right_on="Fecha")#.drop(columns=['Operator','Código Distribuidor','Version'])
demand_power_df = demand_power_df [['Fecha','Region','Demand', 'Demanda Máxima de Potenca kW']]
demand_power_df.columns = ['Date','Region','Energy Demand', 'Maximum demand']
demand_power_df.head()
```
